### Reducing Categories 

```python
income_raw_df['Workclass'].replace(['State-gov','Local-gov', 'Federal-gov'], 'Government', inplace=True)

income_raw_df['Workclass'].replace(['Self-emp-not-inc','Self-emp-inc'], 'Self-Employment', inplace=True)

income_raw_df['Workclass'].replace(['Without-pay','Never-worked'],'Unemployment', inplace=True)
```

```python
income_raw_df['Martial_Status'].replace(['Married-civ-spouse', 'Married-spouse-absent','Married-AF-spouse'],
                                        'Married', inplace=True)

income_raw_df['Martial_Status'].replace(["Divorced","Separated", "Widowed"], "Not Married", inplace=True)
```

```python
income_raw_df["Education"].replace(['11th', '9th', '7th-8th','Prof-school','5th-6th',
                                    '10th', '1st-4th', 'Preschool', '12th'], "HS-dropout", inplace=True)

income_raw_df["Education"].replace([ 'Some-college','Assoc-acdm', 'Assoc-voc'], "Others", inplace=True)
                                   
```

```python
income_raw_df['Country'].replace(['China','Hong','India','Iran','Cambodia','Japan', 'Laos' ,
          'Philippines' ,'Vietnam' ,'Taiwan', 'Thailand'], 'Asia', inplace=True)

income_raw_df['Country'].replace(['Canada','United-States','Puerto-Rico'], 'North_America', inplace=True)

income_raw_df['Country'].replace(['England' ,'France', 'Germany' ,'Greece','Holand-Netherlands','Hungary',
            'Ireland','Italy','Poland','Portugal','Scotland','Yugoslavia'], 'Europe', inplace=True)

income_raw_df['Country'].replace(['Columbia','Cuba','Dominican-Republic','Ecuador',
                             'El-Salvador','Guatemala','Haiti','Honduras',
                             'Mexico','Nicaragua','Outlying-US(Guam-USVI-etc)','Peru',
                            'Jamaica','Trinadad&Tobago', "South"], 'Latin.and.South.America', inplace=True)
```

```python
# Encode the categorical features as numbers
def number_encode_features(df):
    result = df.copy()
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column])
    return result, encoders

# Calculate the correlation and plot it

encoded_data, _ = number_encode_features(income_raw_df)
plt.subplots(figsize=(15, 12))
sns.heatmap(encoded_data.corr(), vmax=.8,annot=True,cmap="Spectral", square=True, center=0)
plt.show()
```

```python
income_processed_df = income_raw_df
```

```python
dummy_columns = ['Workclass', 'Martial_Status', 'Occupation', 'Relationship', 'Race', 'Sex', 'Country']

data_new = income_processed_df

for cat in dummy_columns:
    dummy_cols = pd.get_dummies(income_processed_df[cat], prefix=cat, drop_first=True)
    data_new = data_new.join(dummy_cols)
```

```python
data_vars = data_new.columns.values.tolist()
to_keep = [i for i in data_vars if i not in dummy_columns]
to_keep
```

```python
final_income_data = data_new[to_keep]
final_income_data
```

```python
def encode_cat(col_names, df):
    
    label_encoder_y = LabelEncoder()
    
    for col in col_names:
        df[col] = label_encoder_y.fit_transform(df[col])
        
```

### Model Building

```python
y = ['Income_Class']
X = [i for i in final_income_data.columns.values.tolist() if i not in y]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(final_income_data[X], final_income_data[y], test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
num_cols = ['Age', 'fnlwgt', 'Education-Num', 'Capital_Gain', 'Capital_Loss', 'Hours_per_week']

logistic_numerical = LogisticRegression()

logistic_numerical.fit(X_train[num_cols], y_train)
y_pred = logistic_numerical.predict(X_test[num_cols])
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logistic_numerical.score(X_test[num_cols], y_test)))
```

```python
logistic_lasso = LogisticRegression(penalty='l1')

logistic_lasso.fit(X_train, y_train)
y_pred = logistic_lasso.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logistic_lasso.score(X_test, y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier

decisionTree = DecisionTreeClassifier(criterion = 'entropy', max_depth=7)
decisionTree.fit(X_train, y_train)
y_pred = decisionTree.predict(X_test)
print('Accuracy of decision tree classifier on test set: {:.2f}'.format(decisionTree.score(X_test, y_test)))
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

seed = 7

models = []
models.append(('LR', LogisticRegression(penalty='l1')))
models.append(('DecisionTree', DecisionTreeClassifier(criterion = 'entropy', max_depth=7)))
models.append(('SVM', LinearSVC(C=0.005)))
models.append(('RF', RandomForestClassifier(n_estimators = 35, criterion = 'entropy')))
models.append(('XGBoost', XGBClassifier(max_depth=4, n_estimators=157, learning_rate=0.025)))

results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f" % (name, cv_results.mean()*100)
    print(msg)
```

```python
from sklearn.model_selection import GridSearchCV

parameters={'min_samples_split' : range(10,500,20),'max_depth': range(1,20,2)} 
clf_tree=DecisionTreeClassifier() 
clf=GridSearchCV(clf_tree,parameters) 
clf.fit(X_train,y_train)
clf.best_params_
```

```python
decision_tree_best_fit = DecisionTreeClassifier(criterion='entropy', max_depth=15, min_samples_split=100)
decision_tree_best_fit.fit(X_train, y_train)
y_pred = decision_tree_best_fit.predict(X_test)
print('Accuracy of decision tree classifier on test set: {:.2f}'.format(decision_tree_best_fit.score(X_test, y_test)))
```

```python
importances = pd.DataFrame({'feature':X_train.columns,'importance':np.round(decision_tree_best_fit.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances = importances[importances.values > 0.001]

importances.plot.bar()
```
